```python
import pickle
import sympy as sp

print("Loading results from pickle files...")

with open("moment1.pkl", "rb") as f:
    res1 = pickle.load(f)
    E_x = res1["E_x"]
    t1 = res1["time_seconds"]
    print(f"Loaded E_x: {E_x}")
    print(f"Time taken (Moment 1): {t1:.4f}s")

with open("moment2.pkl", "rb") as f:
    res2 = pickle.load(f)
    sum_weighted_score = res2["sum_weighted_score"]
    t2 = res2["time_seconds"]
    print(f"Loaded Sum Weighted Score: {sum_weighted_score}")
    print(f"Time taken (Moment 2): {t2:.4f}s")
```

```python
# Calculate Variance and Std Dev
Var_x = sum_weighted_score + E_x - E_x**2
Std_Dev = sp.re(Var_x**0.5)

print("\n--- Final Results ---")
print(f"Expectation (E_x): {E_x}")
print(f"Variance (Var_x):  {Var_x}")
print(f"Std Dev:           {Std_Dev}")
print(f"Total Calculation Time: {t1 + t2:.4f}s")
```
